Imputing missing values

```python
df['LoanAmount'].value_counts()
```

```python
for column in ['Gender', 'Dependents', 'Married', 'Education', 'Self_Employed', 'Loan_Amount_Term', 'Credit_History']:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(np.nanmedian(df['LoanAmount']))
```

```python
df.isnull().sum()
```

Label Encoding

```python
df['Education'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for column in ['Gender', 'Dependents', 'Married', 'Education', 'Self_Employed', 'Credit_History']:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['Loan_Status']= encoder.fit_transform(df['Loan_Status'])
```

```python
df.head()
```

One hot encoding

```python
df['Property_Area'].value_counts()
```

```python
dummy_data = pd.get_dummies(df['Property_Area'])
df = pd.concat([df, dummy_data], axis=1)
df.drop(['Property_Area'], axis=1, inplace=True)
```

```python
df.head()
```

Train and Test split

```python
df.drop(['Loan_ID'], axis=1, inplace=True)  # Drop 'Loan_ID' column
y = df['Loan_Status']  # Assign 'Loan_Status' to y
print(y)
X = df[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Rural', 'Semiurban', 'Urban']]  # Assign multiple columns to X
print(X)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=30)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=svc.predict(X_train)
accuracy_score(y_pred,y_train)
```

```python
y_pred=svc.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train,y_train)
```

```python
y_pred=LR.predict(X_train)
accuracy_score(y_pred,y_train)
```

```python
y_pred=LR.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
import numpy as np

def predict_loan_approval(model):
    # Input features
    gender = input("Gender (Male/Female): ").strip().lower()
    married = input("Married (Yes/No): ").strip().lower()
    dependents = input("Dependents: ").strip()
    education = input("Education (Graduate/Not Graduate): ").strip().lower()
    self_employed = input("Self Employed (Yes/No): ").strip().lower()
    applicant_income = float(input("Applicant Income: "))
    coapplicant_income = float(input("Coapplicant Income: "))
    loan_amount = float(input("Loan Amount: "))
    loan_amount_term = float(input("Loan Amount Term (in months): "))
    credit_history = float(input("Credit History (1 for Yes, 0 for No): "))
    rural = input("Rural (Yes/No): ").strip().lower()
    semiurban = input("Semiurban (Yes/No): ").strip().lower()
    urban = input("Urban (Yes/No): ").strip().lower()

    # Convert categorical inputs to binary
    gender_male = 1 if gender == 'male' else 0
    married_yes = 1 if married == 'yes' else 0
    education_graduate = 1 if education == 'graduate' else 0
    self_employed_yes = 1 if self_employed == 'yes' else 0
    rural_yes = 1 if rural == 'yes' else 0
    semiurban_yes = 1 if semiurban == 'yes' else 0
    urban_yes = 1 if urban == 'yes' else 0

    # Prepare feature array
    features = np.array([
        [gender_male, married_yes, dependents, education_graduate, self_employed_yes,
         applicant_income, coapplicant_income, loan_amount, loan_amount_term, credit_history,
         rural_yes, semiurban_yes, urban_yes]
    ])

    # Make prediction
    prediction = model.predict(features)

    if prediction == 1:
        print("Congratulations! Your loan is approved.")
    else:
        print("Sorry, your loan is not approved.")

# Usage example:
# Assuming you have a trained model named 'model'
# predict_loan_approval(model)
```
